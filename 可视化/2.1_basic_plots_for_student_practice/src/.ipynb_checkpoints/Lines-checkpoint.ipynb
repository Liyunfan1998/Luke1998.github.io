   ### 折线图

```python
import numpy as np
import matplotlib.pyplot as plt
 

t = np.arange(0.0, 2.0, 0.05)
s = 1 + np.sin(2*np.pi*t)

# plt.plot(t,s) --- > 换做如下：
line, = plt.gca().plot(t, s)
line.set_marker('*')
line.set_linestyle(':')
line.set_color('r')

plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('About as simple as it gets, folks')
plt.grid(True)
#plt.savefig("test.png")
plt.show()
```
